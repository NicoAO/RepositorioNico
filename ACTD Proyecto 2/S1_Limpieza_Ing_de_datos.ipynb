```python
import pandas as pd
dataFrame = pd.read_excel('default of credit card clients.xls')
dataFrame
```

```python
# Buscar duplicados
dataFrame.duplicated().sum() #Respuesta: 0
```

```python
#Identificar errores estructurales
dataFrame.isnull().values.any()     #Anwser: Flase
```

```python
#Cambiar los nombres de las columnas y eliminar la primera fila que incluia estos nombres.
# Asignar la primera fila como nombres de columna
dataFrame.columns = dataFrame.iloc[0]

# Eliminar la primera fila (que ahora es la fila de nombres de columna)
DataFrame = dataFrame.drop(0)

# Restablecer los índices después de eliminar la primera fila
DataFrame = DataFrame.reset_index(drop=True)

DataFrame
```

```python
import numpy as np

# Obtener los tipos de datos de todas las columnas
tipos_de_datos = DataFrame.dtypes

# Verificar si todos los tipos de datos son numéricos
todos_numericos = tipos_de_datos.apply(lambda x: np.issubdtype(x, np.number)).all()

if todos_numericos:
    print("Todos los valores en la tabla son numéricos.")
else:
    print("Hay columnas que contienen valores no numéricos.")
    print("Tipos de datos por columna:")
    print(tipos_de_datos)
```

```python
# Convertir todas las columnas a tipos de datos numéricos
DataFrame = DataFrame.apply(pd.to_numeric, errors='coerce')

# Verificar los tipos de datos después de la conversión
print(DataFrame.dtypes)

DataFrame
```

```python
DataFrame.to_csv("datosproyecto2")  #  Guardamos el dataframe en un archivo csv
```

```python
import matplotlib.pyplot as plt
numeric_cols = DataFrame.select_dtypes(include=[np.number]).columns
fig, ax = plt.subplots(figsize=(10, 4)) 
ax.boxplot(DataFrame[numeric_cols].values)
ax.set_xticklabels(numeric_cols)
ax.set_ylabel('Value')
plt.show() 
#Ahora --> Hacer una para cada variable para que todo esté en escala
```
