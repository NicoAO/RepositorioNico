```python
import pandas as pd
dataFrame = pd.read_excel('default of credit card clients.xls')
dataFrame
```

```python
# Buscar duplicados
dataFrame.duplicated().sum() #Respuesta: 0
```

```python
#Identificar errores estructurales
dataFrame.isnull().values.any()     #Anwser: Flase
```

```python
#Cambiar los nombres de las columnas y eliminar la primera fila que incluia estos nombres.
# Asignar la primera fila como nombres de columna
dataFrame.columns = dataFrame.iloc[0]

# Eliminar la primera fila (que ahora es la fila de nombres de columna)
DataFrame = dataFrame.drop(0)

# Restablecer los índices después de eliminar la primera fila
DataFrame = DataFrame.reset_index(drop=True)

DataFrame
```

```python
import numpy as np

# Obtener los tipos de datos de todas las columnas
tipos_de_datos = DataFrame.dtypes

# Verificar si todos los tipos de datos son numéricos
todos_numericos = tipos_de_datos.apply(lambda x: np.issubdtype(x, np.number)).all()

if todos_numericos:
    print("Todos los valores en la tabla son numéricos.")
else:
    print("Hay columnas que contienen valores no numéricos.")
    print("Tipos de datos por columna:")
    print(tipos_de_datos)
```

```python
# Convertir todas las columnas a tipos de datos numéricos
DataFrame = DataFrame.apply(pd.to_numeric, errors='coerce')

# Verificar los tipos de datos después de la conversión
print(DataFrame.dtypes)

DataFrame
```

```python
from scipy import stats
numeric_cols = DataFrame.select_dtypes(include=[np.number])
arr = numeric_cols.to_numpy()
z = np.abs(stats.zscore(arr))
outliers = DataFrame[z > 3]
outliers 
### Dejemos los outliers y luego identificar por qué suceden
```

```python
outliers.describe() #7819 outliers identificados
```

```python
ID = []
for i in range(len(outliers)):
    ID.append(outliers.iloc[i,0])
frecuencias = {}
for i in range(len(ID)):
    if ID[i] in frecuencias:
        frecuencias[ID[i]]+=1
    else:
        frecuencias[ID[i]]=1
print(frecuencias)

primero = max(frecuencias, key=frecuencias.get)
print(f"El ID que más aparece en los outliers: {primero}")  

segundo = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)[1]
print(segundo)
tercero = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)[2]
print(tercero)
cuarto = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)[3]
print(cuarto)
```

```python
import matplotlib.pyplot as plt
eq = frecuencias.keys()
fr = frecuencias.values()
fig,ax = plt.subplots()
ax.bar(eq, fr)
ax.set_title("Frecuencias de ID en outliers")
ax.set_ylabel("Frecuencia")
ax.set_xlabel("ID")
colors = ["darkcyan", "cadetblue", "cyan"]
ax.bar(eq, fr, color=colors)
fig.tight_layout
plt.show()
```
