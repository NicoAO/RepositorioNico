Generación de histograma

```python
import matplotlib.pyplot as plt
```

```python
count, bins, ignored = plt.hist(x=df["team"], bins=30)
plt.title('Histograma de numero de equipo asociado con la instancia')
plt.xlabel('Numero equipo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["targeted_productivity"], bins=30)
plt.title('Histograma de productividad objetivo establecida para cada equipo para cada día')
plt.xlabel('productividad establecida')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["targeted_productivity"], bins=30)
plt.title('Histograma de el tiempo asignado para una tarea')
plt.xlabel('Valor de minutos estándar')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["wip"], bins=30)
plt.title('Histograma del trabajo en progreso')
plt.xlabel('Cantidad de trabajo en progreso')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["over_time"], bins=30)
plt.title('Histograma de la cantidad de tiempo extra de cada equipo en minutos')
plt.xlabel('Tiempo en minutos')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["incentive"], bins=30)
plt.title('Histograma de la cantidad de incentivo financiero (en unidades monetarias)')
plt.xlabel('incentivo financiero (en unidades monetarias)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["idle_time"], bins=30)
plt.title('Histograma de La cantidad de tiempo durante el cual la producción estuvo interrumpida')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["idle_men"], bins=30)
plt.title('Histograma del número de trabajadores que estaban inactivos debido a la interrupción de la producción')
plt.xlabel('Número de trabajadores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["no_of_style_change"], bins=30)
plt.title('Histograma de número de cambios en el estilo de un producto en particular')
plt.xlabel('Número de cambios')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["no_of_workers"], bins=30)
plt.title('Histograma de número de trabajadores en cada equipo')
plt.xlabel('Número de trabajadores por equipo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
count, bins, ignored = plt.hist(x=df["actual_productivity"], bins=30)
plt.title('Histograma de número del % real de productividad que entregaron los trabajadores. Varía de 0 a 1')
plt.xlabel('% real de productividad')
plt.ylabel('Frecuencia')
plt.show()
```

Diagrmas de caja

```python
df.boxplot()

plt.title('Diagrama de Caja para cada variable')
plt.xlabel('Variable')
plt.ylabel('Valores')

plt.show()
```

```python
import numpy as np
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(nrows=len(numeric_cols), figsize=(6, 19))
for i, col in enumerate(numeric_cols):
    df[col].plot.box(ax=axes[i])
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```
